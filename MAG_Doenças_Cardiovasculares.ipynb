&nbsp;

Doenças cardiovasculares formam um grupo de disfunções ocorridas no coração e vasos sanguíneos, sendo as principais: 

* **Doença arterial coronária**: inclui angina estável e instável, enfarte do miocárdio e paragem cardiorrespiratória.
* **Doença cerebrovascular**: ligada aos vasos sanguíneos que fornecem sangue ao cérebro, resultando em derrame cerebral ou mesmo Acidente Vascular Cerebral (AVC) hemorrágico.
* **Doença reumática cardíaca**: danos permanentes nas válvulas cardíacas causadas por febre reumática.
* **Doença cardíaca congênita**: má-formação da estrutura cardíaca no nascimento.
* **Trombose e embolia pulmonar**: onde coágulos podem se deslocar das veias da perna e afetar diretamente o pulmão e coração. 

&nbsp;

> # **Estima-se que 17,9 milhões de pessoas morreram por doenças cardiovasculares em 2016, representando 31% de todas as mortes ocorridas no mundo no ano.**

* Obesidade;

Percebe-se que muitas são as variáveis que exercem influência direta ou indireta na manifestação das doenças cardíacas em diversos perfis de indivíduos espalhados pelo mundo, algo que torna a prevenção um considerável desafio.

### **4.1) Identificação e descrição das 14 variáveis (colunas) que compõem o *dataset*** 

* `age` - idade em anos
* `sex` - sexo do paciente
  * 0: mulher
  * 1: homem
* `cp` - tipo da dor torácica
  * 1: angina típica
  * 2: angina atípica
  * 3: dor não cardíaca
  * 4: assintomática
* `trestbps` - pressão arterial em repouso
* `chol` - colesterol sérico (*mg/dl*)
* `fbs` - açucar no sangue em jejum > 120*mg/dl*
  * 0: `False`
  * 1: `True`
* `restecg` - resultado do eletrocardiograma de repouso
  * 0: normal
  * 1: anormalidades de ST-T  (inversão da onda T e elevação ou depressão de > 0.05*mV*)
  * 2: hipertrofia ventricular esquerda provável ou definitiva (pelos critérios de Romhilt-Estes)
* `thalach` - frequência cardíaca máxima atingida
* `exang` - angina induzida pelo exercício
  * 0: não
  * 1: sim
* `oldpeak` - depressão do segmento ST induzida pelo exercício em relação ao repouso
* `slope` - inclinação do segmento ST no pico do exercício
* `ca` - número de vasos principais colorido por fluoroscopia
* `thal` - *thallium stress test* (teste de estresse por tálio)
  * 3: normal
  * 6: defeito fixo
  * 7: defeito reversível
* `num` - diagnóstico de doença cardíaca (diagnóstico angiográfico)
  * 0: não há doença cardíaca ( < 50% de estreitamento do diâmetro)
  * 1,2,3,4: há doença cardíaca ( > 50% de estreitamento do diâmetro)

### **6.1) Distribuição das variáveis e detecção de valores discrepantes (*outliers*) no *dataset***

Para melhor observação sobre a distribuição dos dados, foram plotados histogramas para cada uma das variáveis do *dataset*.



```
df.hist(figsize=(16,12), color='g', grid=False)
plt.tight_layout();
```

Após uma análise superficial dos histogramas referentes a cada uma das variáveis do *dataset*, a distribuição nas variáveis `oldpeak`e `chol` podem apresentar valores discrepantes (*outliers*)

Para nos aprofundarmos nessa questão, vamos primeiro plotar novamente os histogramas das duas variáveis em um tamanho maior: 

```
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(14,4))

df['chol'].plot(kind='hist', ax=ax[0], color='g', grid=False)
ax[0].set_title('Histograma de colesterol sérico')

df['oldpeak'].plot(kind='hist', ax=ax[1], color='g', grid=False)
ax[1].set_title('Histograma de old peak')

sns.set(style='darkgrid')
plt.tight_layout()
```

A distribuição das duas variáveis sugere um pequeno número de ocorrência de valores discrepantes, também conhecidos como *outliers*. 

Para melhor detalhamento da ocorrência desses *outliers*, utilizaremos nas variáveis `chol`e `oldpeak` o gráfico apropriado para isso: *boxplot* 

```
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8,6))

df.boxplot(['chol'], ax=ax[0], widths=0.6, grid=False)
ax[0].set_title('Boxplot Colesterol Sérico', fontsize=12)

df.boxplot(['oldpeak'], ax=ax[1], widths=0.6, grid=False)
ax[1].set_title('Boxplot old peak', fontsize=12)



plt.tight_layout();
```

> * De fato, ambas as variáveis apresentam *outliers*. Cada bola presente no *boxplot* representa um valor muito distante da maioria dos demais valores presentes em cada uma das variáveis.

> * *Outliers* são valores discrepantes que afetam a média de uma variável, podendo reduzir a precisão das análises do *dataset*. Neste caso, considerando que a quantidade de valores discrepantes é baixa nas duas variáveis, os mesmos podem ser excluídos.






### **6.2) Análise de correlação entre as variáveis**

Correlação significa que existe uma relação entre duas coisas. No nosso contexto, buscamos compreender como as 14 variáveis do *dataset* relacionam-se umas com as outras.

Essa relação pode ser medida, e é função do coeficiente de correlação estabelecer qual a intensidade dela. O coeficiente de valor **1** representa  máxima correlação entre duas variáveis.

**Para identificar as correlações existentes entre as variáveis de interesse, seguiremos dois passos:**

*1 - Criação de uma matriz de correlação*

```
df.corr()
```

*2 - Para melhor visualização da matriz de correlação, a aplicaremos em um mapa de calor (heatmap) onde os quadrados com as **tonalidades mais escuras de azul**, representam **correlações positivas** mais representativas; já os quadrados mais **próximos à cor branca**, representam as maiores **correlações negativas**.*

```
corr = df.corr()

f, ax = plt.subplots(figsize=(14, 12))
sns.heatmap(corr, annot=True, linewidths=.5, cmap="YlGnBu")
plt.tight_layout();
```

> O dataset **não apresenta correlações consideradas fortes**, ou seja, **superiores a 0.7** no caso das correlações positivas e **inferiores a -0.7** no caso das correlações negativas.  

> Destaca-se entre as principais **correlações negativas** (o aumento de uma variável implica na diminuição da outra):
* `thalach` e `num` com o nível de correlação: **-0.42**

> Já nas **correlações positivas** (onde o aumento de uma variável está relacionado ao aumento da outra) destacam-se:
* `oldpeak` e `slope` com o nível de correlação: **0.58**
* `thal` e `num` com o nível de correlação: **0.52**

### **6.4) Análise de diagnóstico de doença cardíaca** 

Na base de dados estudada, a variável `num` está relacionada com o diagnóstico angiográfico, classificando se os pacientes analisados foram ou não diagnosticados com doença cardiovascular. 

* O valor **0** representa estreitamento do diâmetro arterial inferior a 50%, o que significa que o paciente **não possui doença cardiovascular.**

* Já os valores **1,2,3,4** representam estreitamentos superiores a 50%, o que significa que o paciente **possui doença cardiovascular.** 




```
# variável alvo antes
df.num.value_counts()
```

Ou seja, apesar de termos 5 possíveis valores, os resultados podem ser somente 2: ou o **indivíduo possui doença cardíaca** ou o **indivíduo não possui doença cardíaca**. Sendo assim, simplificaremos a as classificações da variável `num` da seguinte forma:

* Todos os valores **0** serão mantidos e os valores **1,2,3,4** serão agrupados para **1**.


```
# converter valores >= 1 em 1
df.loc[df.num != 0, 'num'] = 1
```

```
# variável alvo depois
df.num.value_counts()
```

```
print('Não há doença cardíaca:\t{}'.format(df.num.value_counts()[0]))
print('Há doença cardíaca:\t{}'.format(df.num.value_counts()[1]))
```

```
df.num.value_counts(ascending=False) /df.shape[0]
```

> Após a conversão, as análises da variável `num` tornaram-se mais diretas e concisas, onde podemos inferir facilmente que, dos **303** pacientes analisados no *dataset*:
* **164** não possuem doença cardíaca (**54%** do total)
* **139** possuem doença cardíaca (**46%** do total)

### **6.6) Análise da variável `restecg` (Eletrocardiograma em repouso)**

<center><img src="https://www.publicdomainpictures.net/pictures/50000/velka/heart-cardiogram.jpg" width="640" /></center>

De acordo com a Sociedade Beneficente Israelita Brasileira Albert Einstein, o eletrocardiograma em repouso tem como objetivos:

* Analisar o ritmo cardíaco;
* Avaliar a condução do estímulo através do sistema de condução do coração e das suas cavidades;
* Avaliar a integridade ou anormalidades do sistema de condução;
* Detectar eventuais sobrecargas das cavidades cardíacas e zonas correspondentes à ausência de atividade elétrica.

Os resultados desse tipo de procedimento oferecem informações cruciais sobre as atividades cardíacas de um paciente, informações essas que também podem oferecer importantes *insights* na análise de dados. 

Na base de dados da *Cleveland Clinic Foundation*, objeto principal deste estudo, **os resultados do eletrocardiograma em repouso dos pacientes está na variável `restecg`**. 

Sendo assim, vamos plotar os resultados por meio de um **gráfico de pizza**.



```
df['restecg'].value_counts(ascending=False)
```

```
restecg = dict(df.restecg.value_counts())

labels = restecg.keys()
sizes = restecg.values()
explode = (0.1, 0.05, 0.05)
```

```
restecg
```

```
fig, ax = plt.subplots(figsize=(6,6))

ax.pie(sizes, labels=labels, explode=explode, shadow=True )
ax.set_title("Eletrocardiograma em Repouso - Resultados", fontsize=14)
plt.tight_layout()
```

Para melhor visualização dos resultados, vamos substituir o gráfico de pizza por um gráfico de contagem de valores (*countplot*)





```
fig, ax=plt.subplots(figsize=(8,8))
ax = sns.countplot(y=df['restecg'])
ax.set_title("Eletrocardiograma de Repouso (restecg)", fontsize=18)
ax.set_ylabel("Resultados", fontsize=14)
ax.set_xlabel("Número de pacientes", fontsize=14)
ax.set_yticklabels(["Normal", "Anormalidades", "Hipertrofia vent"])
plt.tight_layout;
```

```
print('Normal:\t{}'.format(df['restecg'].value_counts()[0]))
print('Anormalidades de ST-T:\t{}'.format(df['restecg'].value_counts()[1]))
print('Hipertrofia ventricular esquerda provável ou definitiva:\t{}'.format(df['restecg'].value_counts()[2]))
```

### **6.7) Relação entre as variáveis de Colesterol Sérico (`chol`) e idade dos pacientes (`age`)**

Os níveis séricos de colesterol, representados pela variável `chol`, podem balizar o fator de risco do desenvolvimento de uma doença cardiovascular.
Vamos analisar se há uma relação do nível de colesterol sérico, conforme o avanço da idade dos pacientes do *dataset*, representada aqui pela variável `age`.

Para melhor análise dessa relação, utilizaremos um gráfico de dispersão (*scatterplot*) juntamente com uma linha de regressão linear.

```
fig, ax = plt.subplots(figsize=(10,6))
sns.regplot(y='age', x='chol', line_kws={"color": "red"}, data=df, ax=ax)
ax.set_title('Nível sérico de colesterol X Idade', fontsize=16)
ax.set_ylabel('Idade', fontsize=14)
ax.set_xlabel('Colesterol sérico', fontsize=14)
plt.tight_layout()
```

> *Pela análise do gráfico, pode-ser perceber que há uma relação positiva entre a elevação dos níveis séricos de colesterol conforme o avanço da idade dos pacientes avaliados. Ambos são fatores de risco e ambos se relacionam.*
